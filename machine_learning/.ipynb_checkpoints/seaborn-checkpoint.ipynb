### 最一般的使用方法（以factorplot为例）

seaborn.factorplot(x=None, y=None, hue=None, data=None, row=None, col=None, col_wrap=None, estimator=\<function mean>, ci=95, n_boot=1000, units=None, order=None, hue_order=None, row_order=None, col_order=None, kind='point', size=4, aspect=1, orient=None, color=None, palette=None, legend=True, legend_out=True, sharex=True, sharey=True, margin_titles=False, facet_kws=None, \*\*kwargs)

- data: DataFrame数据，一个datasets
- x,y,hue：data中的变量（特征）名
- row,col：指定行（列）以哪个变量取值变化
- order：回归函数的次数
- kind：图的类型，point, bar, count, box, violin, strip等
- size：指定图的尺寸
- orient：指定图的方向
- palette：设置配色
- legeng：是否显示图例

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
# define a sin function
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
import seaborn as sns
sinplot()
```

### 双变量分布

```python
import pandas as pd
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
data.shape
```

```python
# 散点图
sns.jointplot(x="x", y="y", data=df)
```

```python
# Hexbin图
x, y = np.random.multivariate_normal (mean, cov, 1000).T
x.shape, y.shape
```

```python
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
# 核密度估计
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

```python
# 也可用kdeplot()画二维的核密度图
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df.x, df.y, ax=ax)
sns.rugplot(df.x, color="g", ax=ax)
sns.rugplot(df.y, vertical=True, ax=ax);
```

```python
# 增加连续等级使图像更平滑
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.x, df.y, cmap=cmap, n_levels=60, shade=True);
```

```python
# 使用jointplot()返回的JointGrid对象，对图像进一步处理
g = sns.jointplot(x="x", y="y", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

## 回归图绘制


```python
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "regression")))
tips = sns.load_dataset("tips")
tips.shape
tips.head(5)
```

### 线性回归

```python
#线性回归主要包含两个函数：regplot和lmplot 后者的特性比后者更丰富
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="size", y="tip", data=tips)
```

```python
# size变量是离散的，对其做回归分析效果不好，因此对其增加一些“抖动”
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.05);
```

```python
# 另一种方法是以每个离散变量点的中心点进行回归
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean)
```

### 非线性模型

```python
# 使用了 Anscombe’s quartet 数据集
anscombe = sns.load_dataset("anscombe")
anscombe.shape
```

```python
anscombe.head(5)
```

```python
# 线性模型应用的局限
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
```

```python
# 指定参数order的值
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```

```python
# 检测离散点 参数robust设为True
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80});
```

```python
# 对于y值是离散的情况，使用逻辑回归
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03)
```

```python
# 非参数回归
sns.lmplot(x="total_bill", y="tip", data=tips,
           lowess=True);
```

```python
# residplot() 可以用来检验数据是否适用于线性回归，通过画出每个数据点的残差
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
              scatter_kws={"s": 80});
```

```python
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
              scatter_kws={"s": 80});
```

## 分类图绘制
seaborn的分类图可以分为三种：
- 将分类变量的每个级别的每个观察结果显示出来(swarmplot，stripplot)
- 显示每个观察分布的抽象表示(boxplot, violinplot)
- 应用统计估计显示中心趋势度量和置信区间(barplot, pointplot)

以上的函数对于seaborn来说，都是属于较底层的接口，只能画某一种类型的图，factorplot()函数则可以根据数据特征画出各种类型的图

## 数据感知（data-aware）图绘制

当探索维度适中的数据时，一个有效的方法是画出数据子集的多个实例，这种方法称为栅格图（lattice）或网格图（trellis）。该方法要求数据必须是pandas的DataFrame，即每列为变量，每行为变量的观察值。

```python
sns.set(style="ticks")
np.random.seed(sum(map(ord, "axis_grids")))
```

### FacetGrid的使用

```python
# 设置好栅格和坐标轴
g = sns.FacetGrid(tips, col="time")
```

```python
# 调用map函数，提供画图函数和变量名
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist, "tip")
```

```python
# 更多灵活的用法
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1);
```

```python
# 设置尺度
g = sns.FacetGrid(tips, col="day", size=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill");
```

```python
# 设置不同的配色方案
pal = dict(Lunch="seagreen", Dinner="gray")
g = sns.FacetGrid(tips, hue="time", palette=pal, size=5)
g.map(plt.scatter, "total_bill", "tip", s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.add_legend();
```

```python
# 通过 hue_kws设置marker
g = sns.FacetGrid(tips, hue="sex", palette="Set1", size=5, hue_kws={"marker": ["^", "v"]})
g.map(plt.scatter, "total_bill", "tip", s=100, linewidth=.5, edgecolor="white")
g.add_legend()
```

```python
# 调用map函数后还可设置其他特性
with sns.axes_style("white"):
    g = sns.FacetGrid(tips, row="sex", col="smoker", margin_titles=True, size=2.5)
g.map(plt.scatter, "total_bill", "tip", color="#334488", edgecolor="white", lw=.5);
g.set_axis_labels("Total bill (US Dollars)", "Tip");
g.set(xticks=[10, 30, 50], yticks=[2, 6, 10]);
g.fig.subplots_adjust(wspace=.02, hspace=.02);
```

```python
# 将自定义的函数传递给map函数
def quantile_plot(x, **kwargs):
    qntls, xr = stats.probplot(x, fit=False)
    plt.scatter(xr, qntls, **kwargs)

g = sns.FacetGrid(tips, col="sex", size=4)
g.map(quantile_plot, "total_bill");
```

```python
# 双变量
def qqplot(x, y, **kwargs):
    _, xr = stats.probplot(x, fit=False)
    _, yr = stats.probplot(y, fit=False)
    plt.scatter(xr, yr, **kwargs)

g = sns.FacetGrid(tips, col="smoker", size=4)
g.map(qqplot, "total_bill", "tip");
```

### PairGridd的使用

在PairGrid中， 每一个栅格都被赋予一个不同变量组合，这种图叫做“scatterplot matrix”
FacetGrid和PairGrid的区别：
- FacetGrid的棕横坐标是相同的，每个图画的不同hue的取值
- PairGrid中横纵坐标都是不同变量的组合

PairGrid能快速反映出数据集中两两变量的关系。
使用方法与FacetGrid类似，首先初始化一个网格，然后将画图函数传递给map函数

```python
iris = sns.load_dataset("iris")
g = sns.PairGrid(iris)
g.map(plt.scatter);
```

```python
# 一种典型的用法是将hue设为类别，画出各个特征间的关系
g = sns.PairGrid(iris, hue="species")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```

```python
# 只选取部分特征
g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], hue="species")
g.map(plt.scatter);
```

```python
# 使用pairplot快速浏览数据,返回PairGrid对象进一步处理
g = sns.pairplot(iris, hue="species", size=2.5);
```
